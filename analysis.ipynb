<h3>Comprobacion de tipos de datos</h3>

```python
# Transformacion de dataos a datetime
cyclistic_df.started_at = pd.to_datetime(cyclistic_df.started_at)
cyclistic_df.ended_at = pd.to_datetime(cyclistic_df.ended_at)
```

```python
# Creacion de nuevas columnas hour - ride_minutes, ride_distance
cyclistic_df['hour'] = cyclistic_df.started_at.dt.hour

cyclistic_df['ride_minutes'] = (cyclistic_df['ended_at'] - cyclistic_df['started_at'])
cyclistic_df['ride_minutes'] = pd.to_numeric(cyclistic_df['ride_minutes'])
cyclistic_df['ride_minutes'] = np.round(cyclistic_df['ride_minutes']/6e+10, 2)
```

```python
# Calculo de la distancia de viaje usando la formula
# de Haversine, la cual se usa para calcular la distancia
# entre dos punto geograficos
import math
def haversine(lat1, lon1, lat2, lon2):
  """
    Esta funcion calcula la distancia entre dos puntos
    geograficos, utilizando la formula de Haversine
  """
  rad=math.pi/180
  distance_lat=lat2-lat1
  distance_lon=lon2-lon1
  R=6372.795477598  #radio terrestre
  a=(math.sin(rad*distance_lat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*distance_lon/2))**2
  distancia=2*R*math.asin(math.sqrt(a))
  return round(distancia, 2)

```

```python
cyclistic_df['ride_distance'] = cyclistic_df.apply(lambda d: haversine(d.start_lat, d.start_lng, d.end_lat, d.end_lng), axis=1 )
```

```python
cyclistic_df['month'] = cyclistic_df.started_at.dt.month_name()

cyclistic_df['ride_day'] = cyclistic_df.started_at.dt.day_name()
```

```python
# stadisticas de la duracion de viaje en minutos
cyclistic_df.ride_minutes.describe()
```

Como se puede ver al momento de analizar el dato menor en cuanto a los minutos de viaje tenemos un numero negativo, lo cual nos indicca que la fecha de inicio de la carrera es mayor a la de fin, esto se puede dar por un error al momento de tomar los datos, o algun error referente a la conexion a red de las bicicletas y la empresa.

DEbido a esto para el analisis he decidido eliminar todos los datos cuya fecha de inicio sea mayor a la de fin
